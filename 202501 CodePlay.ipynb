```python
#Shaking Moving Drone

from djitellopy import Tello
import time

tello = Tello()
tello.connect()

battery_level = tello.get_battery()
print(f"Battery Life Percentage: {battery_level}")

distance_cumulative = 0
leap_distance= int(input("Kindly provide a measurement on leaps."))
leaps_taken = int(input("Kindly provide the number of leaps you want to do."))

tello.takeoff()
time.sleep(3)


for x in range(0,leaps_taken):
    distance_cumulative += leap_distance
    time.sleep(1)
    for x in range(2):
        tello.move_forward(distance_cumulative)
        tello.rotate_clockwise(180)
        time.sleep(0.5)

time.sleep(3)
tello.land()
```

```python
#Shaking Moving Drone with XYZ

from djitellopy import Tello
import time

tello = Tello()
tello.connect()

battery_level = tello.get_battery()
print(f"Battery Life Percentage: {battery_level}")

distance_cumulative = 0
leap_distance= int(input("Kindly provide a measurement on leaps."))
leaps_taken = int(input("Kindly provide the number of leaps you want to do."))
speed_default = 20

tello.takeoff()
time.sleep(3)

for x in range(0,leaps_taken):
    distance_cumulative += leap_distance
    time.sleep(1)
    tello.go_xyz_speed(distance_cumulative,0,0,speed_default)
    time.sleep(0.5)
    tello.go_xyz_speed(-distance_cumulative,0,0,speed_default)
    tello.rotate_clockwise(180)
    time.sleep(1)

time.sleep(3)
tello.land()
```

```python
#Shaking Moving Drone with Asynch

from djitellopy import Tello
import time

tello = Tello()
tello.connect()

battery_level = tello.get_battery()
print(f"Battery Life Percentage: {battery_level}")

leaps_taken= int(input("Kindly provide how many leaps you want to do."))
leap_length = int(input("Kindly provide the time you want a leap to take. Be wary."))
speed_default = 30

tello.takeoff()
time.sleep(3)

for x in range(1,leaps_taken+1):
    duration = leap_length*x
    tello.send_rc_control(0, speed_default, 0, 0)
    time.sleep(duration)
    tello.send_rc_control(0, -speed_default, 0, 0)
    time.sleep(duration)
    tello.send_rc_control(0, 0, 0, 0)
    time.sleep(1)

time.sleep(3)
tello.land()
```
